```python
import pandas as pd

df1 = pd.DataFrame({
    'id':[1,2,3,4],
    'name':['raja','rani','rahul','rhea'],
    'age':[25,30,35,40]
})

df2 = pd.DataFrame({
    'id':[3,4,5,6],
    "name":['rhaul','rhea','ravi','reena'],
    'age':[35,40,45,50]
})

df3 = pd.merge(df1,df2,on='id',how='inner')
df3.rename(columns={'name_x':'table_name1'},inplace=True)
df3.rename(columns={'name_y':'table_name2'},inplace=True)
# print(df3)


df4 = pd.merge(df1,df2,on='id',how='left')
df4


df5 = pd.merge(df1,df2,on='id',how='right')
df5


df6 = pd.merge(df1,df2,on='id',how='outer')
df6

df7 = pd.merge(df1,df2,how='cross')
df7

df8 = pd.concat([df1,df2])
df8
```
