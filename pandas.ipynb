```python
import numpy as np

products = []
prices = []

n = int(input("Enter number of products: "))

for i in range(n):
    product = input(f"Enter name of product {i+1}: ")
    price = float(input(f"Enter price of {product}: "))

    products.append(product)
    prices.append(price)

p = np.array(prices)

print("\nProducts and Prices:")
for i in range(n):
    print(products[i], ":", prices[i])

print("\nHighest Price:", p.max())
print("Lowest Price:", p.min())
print("Total Price:", p.sum())
print("Average Price:", p.mean())
print("Standard Deviation:", p.std())
```

```python
import numpy as np
#create a mat 5*5 matrix of random integer(1-90);
mt = np.random.randint(1,90,(5,5));
print(mt)
print("\n")


#find mean of values col and row wise
row_mean = np.mean(mt, axis=1)
print("Row Means:", row_mean)
print("\n")

col_mean = np.mean(mt, axis=0)
print("Column Means:", col_mean)
print("\n")

#replace all even numbers with -1
even_num = np.where(mt % 2 == 0, -1,mt)
print("Modified Matrix:")
print(even_num)
print("\n")

# sort the mat col and row wise w
srt = np.sort(mt, axis=0)
print("Sorted Matrix col wise :")
print(srt)
print("\n")

srt = np.sort(mt, axis=1)
print("Sorted Matrix row wise :")
print(srt)
print("\n")

# find sum of row and col wise seperately
row_sum = np.sum(mt, axis=1)
print("Row Sums:", row_sum)
print("\n")

col_sum = np.sum(mt, axis=0)
print("Column Sums:", col_sum)
print("\n")

#cnt the number of -ve in each column

cnt_negative = np.sum(np.where(even_num < 0, 1, 0), axis=0)
print("Count of Negative Numbers:", cnt_negative)
print("\n")
```

```python
import numpy as np
stock = np.random.randint(1,100,(5,5))
demand= np.random.randint(1,100,(5,5))
# print(stock)
# print("\n")
# print(demand)
# print("\n")

rem_stock = stock - demand
print(rem_stock)
print("\n")

rep_neg_val = np.where(rem_stock < 0, 0,rem_stock)
print(rep_neg_val)
print("\n")

#find
# total shortage
total_shortage = np.sum(np.where(rem_stock < 0, -rem_stock, 0))
print("Total Shortage:", total_shortage)
print("\n")

#products with zero stock
zero_stock_products = np.sum(rep_neg_val == 0, axis=1)
print("Products with Zero Stock:", zero_stock_products)
print("\n")

#sort each row based on remaining stock
sorted_stock = np.sort(rep_neg_val, axis=1)
print("Sorted Stock:")
print(sorted_stock)
print("\n")
```
