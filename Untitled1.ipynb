```python
s = input()
l = len(s)
k = 0
j = l - 1

for i in range(l // 2):
    if s[k] == s[j]:
        k += 1
        j -= 1
    else:
        print(s, "is not palindrome")
        break
else:
    print(s, "is palindrome")

```

```python
str = input()
i=0
j=len(str)-1
while i<j:
  if str[i]==str[j]:
    i+=1
    j-=1
  else:
    print(str,"is not palindrome")
    break
else:
  print(str,"is palindrome")
```
